### Leer una imagen

Las imágenes pueden ser interpretadas como matrices cuyos valores refieren a escalas de grises o colores. En el caso de las escalas de grises, una sola matriz puede definir la imagen. En cambio, cuando se trata de imágenes de colores, son necesarias al menos 3 matrices que indiquen el nivel de cada color (como el caso de RGB, por ejemplo).

Abriremos la clásica imagen de la mano de Röntgen que da nacimiento a las imágenes radiológicas.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open('img/x-ray-rontgen.jpg')

plt.imshow(img)
plt.show()
```

Ahora la imagen queda definida como `img`. `PIL` nos permite operar sobre ella y conocer sus propiedades y características a través de diferentes funciones.

Para más información sobre manipulación de imágenes, se pueden consultar los siguientes links:

* https://www.unioviedo.es/galiano/master/lab01_introduccion_imagen_py.html

* https://matplotlib.org/users/pyplot_tutorial.html

* https://claudiovz.github.io/scipy-lecture-notes-ES/advanced/image_processing/index.html

Se pueden probar todas las funciones escribiendo `img.` y dando `tab` para ver las opciones.

```python
bandas = img.getbands()
size = img.size
bits = img.bits
formato = img.format
colores = img.getcolors()
extremos = img.getextrema()
histograma = img.histogram()

print(size)
```

Así, podremos graficar el histograma utilizando la función `hist` de `matplotlib.pylab`.

```python
from matplotlib.pylab import hist
import matplotlib.pyplot as plt
hist(img.histogram())
plt.show()
```

Una buena herramienta para visualización de datos es `seaborn`, optimizado para usar con `pandas` y analizar datasets o grandes volúmenes de datos.

```python
import seaborn as sns
sns.distplot(img.histogram())
plt.show()
```

### Crear una imagen en python

Una imagen, como vimos, es una matriz. Crear una imagen es, entonces, como crear una matriz.

```python
# ejemplito de como construir un vector de forma abreviada
a = []
for i in range(10):
    a.append(i)
    
b = [[i*2*j for i in range(10)] for j in range(5)]
    
b
```

```python
w, h = 30, 45;
matrix = [[(x-y)**2 for x in range(w)] for y in range(h)]

plt.imshow(matrix)
plt.show()
```

```python
matriztexto = [[str((x,y)) for x in range(w)] for y in range(h)]
matriztexto[44][29]
```
