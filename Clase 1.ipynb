### Leer una imagen

Las imágenes pueden ser interpretadas como matrices cuyos valores refieren a escalas de grises o colores. En el caso de las escalas de grises, una sola matriz puede definir la imagen. En cambio, cuando se trata de imágenes de colores, son necesarias al menos 3 matrices que indiquen el nivel de cada color (como el caso de RGB, por ejemplo).

Abriremos la clásica imagen de la mano de Röntgen que da nacimiento a las imágenes radiológicas.

### Crear una imagen en python

Una imagen, como vimos, es una matriz. Crear una imagen es, entonces, como crear una matriz.

```python
import matplotlib.pyplot as plt

w, h = 30, 45;
matrix = [[(x-y)**2 for x in range(w)] for y in range(h)]

plt.imshow(matrix)
plt.show()
```

También se puede leer la imagen de la mano de Rayos X utilizando `plt` y la librería `misc` de `scipy`.

```python
from scipy import misc
img2 = misc.imread('img/x-ray-rontgen.jpg')
plt.imshow(img2)
```
