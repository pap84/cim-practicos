## Ejercicio 1:

**Derivada primera** de una imagen medica 2D

*Puedo calcular la derivada respecto de el eje x o y*

```python
%matplotlib inline

from matplotlib import pyplot as plt
from PIL import Image
import numpy as np

img = Image.open('../github/img/x-ray-rontgen.jpg')
img= img.convert("L")

img_mat = np.array(img)
```

```python
N=len(img_mat[:,0])  #y
M=len(img_mat[0,:])  #x

img_derx=np.zeros((N,M))
img_dery=np.zeros((N,M))

for i in range(N):
    img_derx[i,:] = np.gradient(img_mat[i,:])

for j in range(M):
    img_dery[:,j] = np.gradient(img_mat[:,j])
    
    
fig = plt.figure(figsize=(15,6))

ax1 = plt.subplot(131)
ax2 = plt.subplot(132)
ax3 = plt.subplot(133)

ax1.imshow(img_mat, cmap='gray', vmin=0, vmax=255)
ax2.imshow(img_derx,  cmap='gray', vmin=0, vmax=255)
ax3.imshow(img_dery,  cmap='gray', vmin=0, vmax=255)

ax1.set_title(u'imagen original', fontsize=14)
ax2.set_title(u'imagen derivada: eje x', fontsize=14)
ax3.set_title(u'imagen derivada: eje y', fontsize=14)

plt.show()
```

## Ejercicio 2:

**Metodo de Sobel** de una imagen medica 2D. Obtener las componentes ∇m y ∇n del operador gradiente.

```python
from scipy import ndimage

mat_m = np.array([[-1,0,1],[-2,0,2],[-1,0,1]]) #x
mat_n = np.array([[-1,-2,-1],[0,0,0],[1,2,1]]) #y

p_sobel = np.mean(img_mat)

grad_m = ndimage.convolve(img_mat, mat_m)
grad_n = ndimage.convolve(img_mat, mat_n)


#extra: (para el ejercicio siguiente) 
#operador gradiente (lo convierto a float32 porque no grafia plt float16) y angulo
grad = np.hypot(grad_m, grad_n).astype(np.float32)
angulo = np.ones((N,M))

for i in range(M):
    for j in range(N):
        if grad_n[j,i]!=0:
            angulo[j,i] = np.arctan(grad_m[j,i]/grad_n[j,i])
        else:
            angulo[j,i] = np.pi/2.
```

## Ejercicio 3:

**Metodo de Sobel** utilizando como parametro de umbralamiento el valor medio de la imagen medica a fin de calcular y graficar la imagen binarizada resultante, la imagen de gradiente |∇| y la imagen de ángulo $\theta$.

```python
#funcion de scipy con la cual chequeear
img_sobel = ndimage.sobel(img_mat)

#resolucion:
sobel = np.zeros((N,M))

for i in range(M):
    for j in range(N):
        if np.abs(grad[j,i])<p_sobel:
            sobel[j,i] = 1

#GRAFICOS:

plt.imshow(img_mat, cmap='gray',vmin=0, vmax=255)
plt.title(u'imagen original', fontsize=14)
plt.show()

fig= plt.figure(figsize=(13,10))

ax1 = plt.subplot(221)
ax2 = plt.subplot(222)
ax3 = plt.subplot(223)
ax4 = plt.subplot(224)

ax1.imshow(sobel, cmap='gray')
ax2.imshow(img_sobel, cmap='gray')
ax3.imshow(grad, cmap='gray')
ax4.imshow(angulo, cmap='gray')


ax1.set_title(u'imagen Sobel: binaria', fontsize=14)
ax2.set_title(u'imagen Sobel: .sobel', fontsize=14)
ax3.set_title(u'imagen Sobel: gradiente', fontsize=14)
ax4.set_title(u'imagen Sobel: angulo', fontsize=14)

plt.show()
```

## Ejercicio 4:

**Método de Prewitt** de una imágen médica 2D.

```python
img_prewitt = ndimage.prewitt(img_mat)

#También se puede hacer usando la función convolución:

matp_m = np.array([[-1,0,1],[-1,0,1],[-1,0,1]]) #x
matp_n = np.array([[-1,-1,-1],[0,0,0],[1,1,1]]) #y

p_prewitt = np.mean(img_mat)

gradp_m = ndimage.convolve(img_mat, matp_m)
gradp_n = ndimage.convolve(img_mat, matp_n)

gradp = np.hypot(grad_m,grad_n)
prewitt = np.zeros((N,M))

for i in range(M):
    for j in range(N):
        if np.abs(gradp[j,i])<p_prewitt:
            prewitt[j,i] = 1

fig = plt.figure(figsize=(17,6))

ax1 = plt.subplot(131)
ax2 = plt.subplot(132)
ax3 = plt.subplot(133)

ax1.imshow(img_mat, cmap='gray', vmin=0, vmax=255)
ax2.imshow(img_prewitt,  cmap='gray', vmin=0, vmax=255)
ax3.imshow(prewitt,  cmap='gray', vmin=0, vmax=1)

ax1.set_title(u'imagen original', fontsize=14)
ax2.set_title(u'imagen Sobel: .prewitt', fontsize=14)
ax3.set_title(u'imagen Sobel: .convolve', fontsize=14)

plt.show()
```
