### Angulo y detector

Quiero saber cual es el ángulo máximo con el cual puede salir un haz de la fuente y ser detectado. Recordar $R_{det}$ esta asociado al tamaño del detector y L a la distancia detector-fuente.

$$ \tan \theta_{det} = \frac{R_{det}}{L}$$

*Por ejemplo, si tomo L=20 y R=5:* 

$$\theta_{det} \sim 0.245 \ rad \sim 14 ^{\circ} $$


Enonces si tomara el eje en el cual uno el centro del detector con la esfera como el eje Z. Tengo que:   $\varphi \in \left[ 0, \ 2 \ \pi \right]$  y $ \theta \in \left[\ 0, \ \theta_{det} \right]$

Sólo me interesan las particulas en ese cono. Luego veo que cantidad son respecto del total de particulas. 


### Cantidad de partículas que llegan al detector y tiempo de adquisición

La cantidad M de partículas que son emitidas en el ángulo que corresponde al detector son sólo un porcentaje del total de partículas N emitidas por la fuente. Están relacionadas de la siguiente manera:

$$ N = \frac{4 \ \pi}{\Omega} M $$

Siendo $\Omega$ el angulo de apertura en estereorradianes. Si quisiera asociarlo con $\theta_{det}$:

$$ \Omega = 2\ \pi \left[ \ 1- \cos \theta_{det} \right]  \left[ sr \right]$$

Luego si tengo M partículas emitidas dentro del ángulo que corresponde al detector:

$$ N = \frac{2}{1- \cos \theta_{det}} M $$

La cantidad de partículas N son emitidas durante un tiempo de exposición $t_e$. Como tengo *10 mCi* entonces:

$$t_{e} = \frac{N}{3,7 x 10^7} \left[ s \right]$$

En función de las variables que intervienen en el programa:

$$t_{e} = \frac{2 \ M}{(1- \cos \theta_{det}) \ 3,7 x 10^7} \left[ s \right]$$


### Relación Intensidad total e Intensidad en angulo sólido de interes:

Cuando quiera trabajar con la intensidad dentro del ángulo sólido de interes tengo que relacionarla con la intensidad de la fuente. Es decir, N esta asociado a la intensidad $I_{o}$ mientras que M esta asociado a $I_{o \Omega}$.

Luego si trabajo con $I_{o \Omega}$ para cada fracción del ángulo del detector cuando quiera realacionarlo con la $I_o$:

$$ I_o = I_{o \Omega} \ \ \frac{2}{1- \cos \theta_{aux}}$$

Siendo en el programa:
$$\theta_{aux} = \frac{\theta_{det}}{div_{haz}}$$

Si para graficar quiero normalizar respecto de $I_o$:
$$ I_{o \Omega} = \frac{1- \cos \theta_{aux}}{2}$$


```python
#IMPORTANTE:
#se pueden variar todos los parametros en el programa menos la cantidad de divisiones en el detector.
#Defini bins de 0,5 cm.
#Datos independientes de R_det y L:


E0= 40 
mu= 0.2683 #Agua 40keV
Io=1.


R_esfera = 5.

#DATOS: L: distancia al detec., R: radio del detec., 
#       haz_div: divisiones del haz, det_div: divisiones del det

for j in range (4):
#PRUEBA 0: L = 20 , R = 5  (L/R = 4)
    if j == 0: 
        L=20.  
        R= 5.
        haz_div = 400
        det_div = (R*2)*2+1
        L0=str(L)
        R0=str(R)
        
#PRUEBA 1:  L = 20 , R = 20 (L/R = 1 )
    if j == 1: 
        L=20. 
        R=20.
        haz_div = 400
        det_div = (R*2)*2+1
        L1=str(L)
        R1=str(R)

#PRUEBA 2:  L = 10 , R = 20 (L/R = 0.5)
    if j == 2: 
        L=10.  
        R=20.
        haz_div = 400
        det_div = (R*2)*2+1
        L2=str(L)
        R2=str(R)
        
#PRUEBA 3:  L = 30 , R = 20 (L/R = 1.5)
    if j == 3: 
        L=100. 
        R= 20.
        haz_div = 400
        det_div = (R*2)*2+1
        L3=str(L)
        R3=str(R)

    tita_det = np.arctan(R/L)  #angulo máximo que cae en el detector

    tita_aux = tita_det/haz_div
    Io_aux = (1.-np.cos(tita_aux))/2.
    If_aux = Io_aux*np.exp(- mu * R_esfera)
    division = int(haz_div*2.+1.)
    y = np.zeros(int(det_div))


    for i in range (division):
        tita = tita_det - tita_aux * i
        y_tita = int(round(L* np.tan(tita),2)*10)
        a = y_tita % 10
        if a == 3 or a == 4 or a==5 or a == 6 or a == 7:
            y_tita = y_tita + (5-a)
        else: 
            y_tita = int(round(y_tita/10.)*10)
        y[int((y_tita+R*10)/5)] += 1

#Guardar datos:
#como tienen cantidad distintas de divisiones guardo en vectores separados:
    if j == 0:
        det0 = y *If_aux
        eje_x0 = ( np.arange(det_div)/2. - R )
        
    if j == 1:
        det1 = If_aux*y
        eje_x1 = ( np.arange(det_div)/2. - R )
        
    if j == 2:
        det2 = If_aux*y
        eje_x2 = ( np.arange(det_div)/2. - R )
        
    if j == 3:
        det3 = If_aux*y
        eje_x3 = ( np.arange(det_div)/2. - R )
```

```python
#IMPORTANTE:
#se pueden variar todos los parametros en el programa menos la cantidad de divisiones en el detector.
#Defini bins de 0,5 cm.
#Datos independientes de R_det y L:

E0= 40 
mu= 0.2683 #Agua 40keV
Io=1.

L=20.  
R= 5.
L0=str(L)
R0=str(R)

haz_div = 400
det_div = (R*2)*2+1
tita_det = np.arctan(R/L)  #angulo máximo que cae en el detector
tita_aux = tita_det/haz_div
Io_aux = (1.-np.cos(tita_aux))/2.
division = int(haz_div*2.+1.)


#Guardar datos:
eje_x= ( np.arange(det_div)/2. - R )    
detector =np.zeros((int(det_div),4))


#DATOS: L: distancia al detec., R: radio del detec., 
#       haz_div: divisiones del haz, det_div: divisiones del det

for j in range (4):
      
    y = np.zeros(int(det_div))

    for i in range (division):
        tita = tita_det - tita_aux * i
        y_tita = int(round(L* np.tan(tita),2)*10)
        a = y_tita % 10
        if a == 3 or a == 4 or a==5 or a == 6 or a == 7:
            y_tita = y_tita + (5-a)
        else: 
            y_tita = int(round(y_tita/10.)*10)
        y[int((y_tita+R*10)/5)] += 1
        
    
#PRUEBA 0: Radio esfera = 5
    if j == 0: 
        R_esfera = 5.
        If_aux=Io_aux*np.exp(- mu * R_esfera)
        R_e0= str(R_esfera)
        detector[:,0] = y *If_aux
#PRUEBA 1: Radio esfera = 10
    if j == 1: 
        R_esfera = 10.
        If_aux=Io_aux*np.exp(- mu * R_esfera)
        R_e1= str(R_esfera)   
        detector[:,1] = y *If_aux
#PRUEBA 2: Radio esfera = 15
    if j == 2: 
        R_esfera = 15.
        If_aux=Io_aux*np.exp(- mu * R_esfera)
        R_e2= str(R_esfera)
        detector[:,2] = y *If_aux
#PRUEBA 3: Radio esfera = 2
    if j == 3: 
        R_esfera = 2.
        If_aux=Io_aux*np.exp(- mu * R_esfera)
        R_e3= str(R_esfera)
        detector[:,3] = y *If_aux
```

```python
#EXTRA: si se quieren los gráficos sin la línea que corresponde a If=Io [eje y=1] 

figure = plt.figure(figsize=(15,20))

ax1 = plt.subplot(221)
ax2 = plt.subplot(222)
ax3 = plt.subplot(223)
ax4 = plt.subplot(224)

ax1.plot(eje_x, detector[:,0])
ax2.plot(eje_x, detector[:,1])
ax3.plot(eje_x, detector[:,2])
ax4.plot(eje_x, detector[:,3])

ax1.set_title(u'L={0},$R_d= ${1}, Radio de la esfera:{2}'.format(L0,R0,R_e0), fontsize=14)
ax2.set_title(u'L={0},$R_d=$ {1}, Radio de la esfera:{2}'.format(L0,R0,R_e1), fontsize=14)
ax3.set_title(u'L={0},$R_d=$ {1}, Radio de la esfera:{2}'.format(L0,R0,R_e2), fontsize=14)
ax4.set_title(u'L={0},$R_d=$ {1}, Radio de la esfera:{2}'.format(L0,R0,R_e3), fontsize=14)

ax1.set_xlabel(u'eje x detector [cm]', fontsize=14)
ax1.set_ylabel(u'Intencidad [If/Io]', fontsize=14)
ax2.set_xlabel(u'eje x detector [cm]', fontsize=14)
ax2.set_ylabel(u'Intencidad [If/Io]', fontsize=14)
ax3.set_xlabel(u'eje x detector [cm]', fontsize=14)
ax3.set_ylabel(u'Intencidad [If/Io]', fontsize=14)
ax4.set_xlabel(u'eje x detector [cm]', fontsize=14)
ax4.set_ylabel(u'Intencidad [If/Io]', fontsize=14)

plt.show()
```

### Variación de los distintos parametros y observaciones:

* **Radio de la esfera:** El radio de la esfera lo único que modifica es la razón entre la Intensidad final e inicial. La forma de las curvas para valores L y $R_d$ no varian mientras se modifica este parámetro. Es más su incidencia en los gráficos sólo se observa si se mira la atenuación (valores eje y).


* **Radio/distancia detector:** Ejemplifique 3 casos distintos:

    1. L >> R : En el primer gráfico la distancia al detector es 4 veces mayor que el Radio de la esfera. Se puede observar que en este caso el espectro se asemeja a lo que se veria con haces paralelos o lo que se espera ver con el uso de un colimador. 

    2. L = R : Se empiezan a observar escalones en los gráficos que son consecuencia de que la esfera emita de forma isotrópica pero radialmente mientras que el detector es plano. La cantidad de haces que son detectados en los 'bins' centrales es mayor que en la parte externa del detector.

    3. L < R : Se observa decaimiento de la intensidad del haz de manera más pronunciada consecuencia de lo dicho anteriormente. 

En resumen, lo óptimo sería trabajar con un detector lo suficientemente alejado de la fuente o utilizar un colimador para corregir/eliminar los haces que no inciden "paralelos".
