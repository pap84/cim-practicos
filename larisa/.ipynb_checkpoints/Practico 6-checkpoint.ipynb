## Penelope
Grafico los resultados obtenidos utilizando el archivos generados: `z-dose.dat` ya que sólo me interesa ver la dosis en profundidad (deposito de energía a lo largo dle eje z). Esto se puede ver en la forma en que definí los bins en el input. 

*Respecto del archivo `z-dose.dat`, (Dose distribution along the central Z axis)*
* *1st column: z (cm).*
* *2nd column: dose (eV/g).*
* *3rd column: statistical uncertainty (3 sigma).*

```python
#importo los datos y los asocio a z y dose
data1=np.loadtxt('z-dose1.dat')  

z1 = data1[:,0]
dose1 = data1[:,1]

#maximo "dose"
max1_E = np.max(dose1)
indice_max = dose1.argmax()
z1_maxE = z1[indice_max]

fig = plt.figure(figsize=(10,6))

p1= plt.plot(z1, dose1, color='red', label='PDD')
#p2= plt.plot(z_maxE, max_E, 'bo', label='maximo deposito de E')
p2= plt.plot(z1_maxE, max1_E, 'bo', label='maximo deposito de E')
p3= plt.plot([z1_maxE, z1_maxE],[0, max1_E], 'b--')
#p4= plt.plot([0,0], [0,400], 'g--', [20,20], [0,400], 'g--', label='limites fantoma')


plt.ylabel('z [cm]', fontsize=11)
plt.xlabel('dose [eV/g]', fontsize=11)
plt.xticks(fontsize=10)
plt.xlim(xmin=-0.5, xmax=20.5)
plt.yticks(fontsize=10)
plt.title('PDD: fantoma cubico de 20cm, fotones de 10 MeV', fontsize=15)
plt.grid(True)
plt.legend()
plt.show()
```

### Comparaciones:

Los electrones a diferencia de los fotones tienen carga, debido a esto tienen más probabilidad de interactuar. Es decir, la sección eficaz del haz de electrones es mayor a la de fotones (y menor camino libre medio).

Esto se ve reflejado en el gráfico, los electrones depositan toda su energía en los primeros centímetros del fantoma mientras que el haz de electrones logra atravesarlo depositando así menor energía en él pero de manera más homogenea.

Entonces se puede ver que no sólo las dosis son distintas sino que también su distribución lo es. En el ambito de la física  médica esto se ve reflejado en el uso de ambos haces. Como los electrones depositan toda su energía en la superficie suelen ser empleados en terapias para la piel, mientras que los fotones al dejar un menor deposito de energía y tener mayor camino libre medio son utilizados para diagnostico.
