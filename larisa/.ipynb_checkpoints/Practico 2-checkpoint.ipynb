## Ejercicio 1:

Estudiar el **algoritmo *fft2*** .

`fft2` es la versión 2D del algoritmo de Fast Fourier Transform (FFT). Se puede encontrar como función dentro de la libreria `numpy` y del módulo `fft`, para poder implementarla (luego de importar np) debo escribir: `np.fft.fft2`.

**Extra: consideraciones a tener en cuenta a la hora de graficar**

* Cuando uso la función `fft2` me devuelve una matriz con números complejos, para poder graficar usando `pyplot` tengo que pasar a números reales.
* Los valores que obtuve aparte de complejos también son muy grandes (ver máximo) necesito reescalearlos, llevarlos a 0-255 o usar una escala logaritmica (depende que este graficando).

```python
%matplotlib inline

from matplotlib import pyplot as plt
from PIL import Image
import numpy as np

img = Image.open('../img/x-ray-rontgen.jpg')
img= img.convert("L")

img_mat = np.array(img)
print ("Tamaño de la matriz:{}".format(img_mat.shape))
```

```python
img_fft2 =np.fft.fft2(img_mat)

print ("Maximo de fft2:{}".format(np.amax(img_fft2.real)),
       "Minimo de fft2:{}".format(round(np.amin(img_fft2.real)),2))

#grafico
fig = plt.figure(figsize=(13,6))

ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

ax1.imshow(img_mat, cmap='gray', vmin=0, vmax=255)
ax2.imshow(img_fft2.real,  cmap='gray', vmin=0, vmax=255)

ax1.set_title(u'imagen original', fontsize=14)
ax2.set_title(u'imagen transformada', fontsize=14)

plt.show()
```

## Ejercicio 2:

Cálculo de distintas **metricas**:

* **Distancia euclidea:** $$  D(k,k',l,l') = \sqrt{(k-k')^2+(l-l')^2} $$
* **Distancia D4:** $$ D(k,k',l,l') = \lvert k-k'\rvert+\lvert l-l'\rvert $$
* **Distancia D8:** $$ D(k,k',l,l') = max(\lvert k-k' \rvert,\lvert l-l'\rvert) $$

**Extra:**
Programa que compara las distintas métricas graficando la imagen sólo para valores menores que una distancia R elegida. *Se espera ver un circulo, rombo y cuadrado.*

```python
def euclidea(imagen, x1, x2, y1, y2):
    DistanciaE = np.sqrt( (x1-x2)**2 + (y1-y2)**2)
    return DistanciaE

def D4(imagen, x1, x2, y1, y2):
    Distancia4 = abs(x1-x2) + abs(y1-y2)
    return Distancia4
    
def D8(imagen, x1, x2, y1, y2):
    Distancia8 = max(abs(x1-x2),abs(y1-y2))
    return Distancia8
```

```python
#Grafico distantas distancias tomandolas <R
R = 70
N=len(img_mat[:,0])  #y
M=len(img_mat[0,:])  #x

if 0==np.mod(N,2):
    N = N-1

if np.mod(M,2)==0:
    M = M-1

c=[int(N/2.),int(M/2.)]

imgr_eu = np.zeros((N,M))
imgr_d4 = np.zeros((N,M))
imgr_d8 = np.zeros((N,M))

#ecuclidea:
for i in range(M):
    for j in range(N):
        dist = euclidea(img_mat, c[1], i, c[0], j)
        if  R>= dist:
            imgr_eu[j,i] = img_mat[j,i]
        else:
            imgr_eu[j,i] = 0

#D4:
for i in range(M):
    for j in range(N):
        dist = D4(img_mat, c[1], i, c[0], j)
        if  R>= dist:
            imgr_d4[j,i] = img_mat[j,i]
        else:
            imgr_d4[j,i] = 0

#D8:
for i in range(M):
    for j in range(N):
        dist = D8(img_mat, c[1], i, c[0], j)
        if  R>= dist:
            imgr_d8[j,i] = img_mat[j,i]
        else:
            imgr_d8[j,i] = 0
```

```python
fig = plt.figure(figsize=(15,10))

ax1 = plt.subplot(221)
ax2 = plt.subplot(222)
ax3 = plt.subplot(223)
ax4 = plt.subplot(224)

ax1.imshow(img_mat, cmap='gray', vmin=0, vmax=255)
ax2.imshow(imgr_eu,  cmap='gray', vmin=0, vmax=255)
ax3.imshow(imgr_d4,  cmap='gray', vmin=0, vmax=255)
ax4.imshow(imgr_d8,  cmap='gray', vmin=0, vmax=255)

ax1.set_title(u'imagen original', fontsize=14)
ax2.set_title(u'imagen con D euclidea <R', fontsize=14)
ax3.set_title(u'imagen con D4 <R', fontsize=14)
ax4.set_title(u'imagen con D8<R', fontsize=14)

plt.show()
```

Algoritmo para calcular **valores medios** en diferentes bloques de una imagen 2D. *Se dividio la imagen en 4 bloques*

## Ejercicio 6:

Algoritmo para aplicar **filtros de suavizado de paso alto/bajo** siendo M×N la dimension de la imagen.

*Teoria:*

Aplicando la definición de transformada de Fourier, se obtiene:
$$ F_{f,g}(m∗, n∗) ≡ TF[f (m, n) ⊗ g(m, n)] = TF[f (m, n)] \,TF[g(m, n)] = F (m∗, n∗) \, G(m∗, n∗) $$

Siendo $G_P(m*,n*)$ la función de transferencia se define como:

* Filto de paso alto:

$$G_P(m∗ , n∗ ) = \left \{  \begin{array}{lcc} 0 \hspace{1.5cm} si \hspace{0.5cm} D(m*,n*) \leq D_{max} \\ 1 \hspace{1.5cm} si \hspace{0.5cm}D(m*,n*) \gt D_{max} \end{array} \right . $$

* Filtro de paso bajo:

$$G_P(m∗ , n∗ ) = \left \{ \begin{array}{lcc} 1 \hspace{1.5cm} si \hspace{0.5cm} D(m*,n*) \leq D_{max} \\ 0 \hspace{1.5cm} si \hspace{0.5cm}D(m*,n*) \gt D_{max} \end{array} \right . $$

```python
def distE(imagen, x2, y2):
    DistanciaE = np.sqrt( x2**2 + y2**2)
    return DistanciaE

N=len(img_mat[:,0])  #y
M=len(img_mat[0,:])  #x

umbral = np.mean(img_fft2)

filtro_alto=np.zeros((N,M))
filtro_bajo=np.zeros((N,M))

#PASO ALTO
for i in range(M):
    for j in range(N):
        if distE(img_fft2,j,i)>umbral:
            filtro_alto[j,i] = 1
            
imgt_alto = np.zeros((N,M))
imgt_alto =img_fft2*filtro_alto
paso_alto = np.fft.ifft2(imgt_alto)
        
#PASO BAJO
for i in range(M):
    for j in range(N):
        if distE(img_fft2,j,i)<=umbral:
            filtro_bajo[j,i] = 1

imgt_bajo = np.zeros((N,M))
imgt_bajo =img_fft2*filtro_bajo
paso_bajo = np.fft.ifft2(imgt_bajo)
```

```python
fig = plt.figure(figsize=(15,6))

ax1 = plt.subplot(131)
ax2 = plt.subplot(132)
ax3 = plt.subplot(133)

ax1.imshow(img_mat, cmap='gray', vmin=0, vmax=255)
ax2.imshow(paso_alto.real,  cmap='gray', vmin=0, vmax=255)
ax3.imshow(paso_bajo.real,  cmap='gray', vmin=0, vmax=255)

ax1.set_title(u'imagen original', fontsize=14)
ax2.set_title(u'imagen suavizada: filtro de paso alto', fontsize=14)
ax3.set_title(u'imagen suavizada: filtro de paso bajo', fontsize=14)

plt.show()
```
