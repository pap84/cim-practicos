* *Definí el scatering simétrico y con mayor probabilidad en 0 y 180, mientras que se acerca a 90 va disminuyendo, esta graficado al final (extra)*
* *Los datos los voy a utilizar para hacer un gráfico 2D y un perfil del deposito de energía*

```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np

#Datos iniciales: (elijo valor máximo de la fuente de liifamir)
E0 = 50.
x0 = 0.
y0 = 0.
tita0 = 0.

tol = 2. #(Cuando corta mi programa)
mat_A = np.zeros((241,230)) #(tamaño de la imagen x-y que muestra los depositos)
vec_E = np.zeros(230) #(vector usado para hacer el perfil energía-profundidad)
k1 = 0 #(contabiliza particulas que salen del medio, i.e, no depositan toda la energia en él)
k2 = 0 #(contabiliza particulas que se van fuera de rango graficado)
k3 = 0 #(para que veas que aunque w = npmax -1 se contabiliza la cantidad de particulas de npmax)
i = -1 #(solo para que la primera vez no anote los datos, sin odepsues de entrar en el loop)

#supongo viaja en vacio hasta llegar al punto x=0, y=0 donde se encuentra con el medio 
E = E0 
x = x0 
y = y0 
tita = tita0

#Calcular depositos de energia/scattering (i.e, interaccion)
#sección eficaz (como la elijo yo a la función la tomo en función de E nomás.. menor E -> menor clm)
```

```python
#Voy a hacer un ajuste de la curva usando la función scipy

from scipy.optimize import curve_fit

#aproximacion con valores dados
def polinomial(z, *grado):
    """z es la variable y grado es el grado del polinomio"""
    poli = 0
    for i, n in enumerate(grado):
        poli += n * z**i
    return poli

import numpy as np
grado = 4
p0 = np.ones(grado + 1)
ejex = np.arange(230)

coeff, var_matrix = curve_fit(polinomial, ejex, vec_E, p0=p0)

yfit = [polinomial(xx, *tuple(coeff)) for xx in ejex]

plt.plot(ejex, vec_E, label='simu')
plt.plot(ejex, yfit, label='fit')
plt.show()
```

## Ejercicio 4: 

Aplicar la situación del item precedente en el caso en que el medio irradiado consiste de un slab (trozo) de material de espesor D a lo largo del eje de incidencia y de dimensiones infinitas a fines prácticos en las direcciones transversales. Asumiendo, que en primera aproximación los efectos de interacción pueden reducirses a eventos de tipo elásticos con desviaciones angulares están determinadas por la función f (cos θ), obtener una expresión a partir de la cual podría calcularse el flujo de energía o partículas.

*Voy a considerar que me interesa saber el flujo cuando salen del slab: z=D, para fines prácticos tomo D=70*
*Cómo sólo considero interacciones elasticas no hay depósito de energía, por lo tanto no hay absorción.*
*Defino la función f(cos $\theta$) igual que en el ejercicio anterior.*
*Si se considerara absorción debería tener en cuenta también la variación de las energías*
*Sólo me importa la disperción de las partículas que salen del material, es decir, con que ángulo*

*El que me interesa es el ángulo: 
* El $\theta$ que es el que me da la simulación, es el parcial con el que sale la partícula 

Los valores finales de posición son $(x_f,y_f)$ mientras que la posición que corresponde a la salida de la partícula del slab corresponde al punto $(D,y_{exit})$ siendo *D* el ancho del slab. Es importante notar el ángulo para estos dos puntos es el mismo $\theta$.

El ángulo $\theta_f$ es el ángulo final que voy a utilizar, ya que el ángulo parcial con el que sale la particula no está definido necesariamente entre (-180,180). 

```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np


#Datos iniciales: (elijo valor máximo de la fuente de liifamir)
E0 = 50.
x0 = 0.
y0 = 0.
tita0 = 0.

D = 70 #(tamaño del slab)

vec_ang = np.zeros(181)
k1 = 0 #(contabiliza particulas que salen del medio)
k2 = 0 #(contabiliza particulas que se van fuera de rango graficado)
i = -1 #(solo para que la primera vez no anote los datos, sin odepsues de entrar en el loop)

#supongo viaja en vacio hasta llegar al punto x=0, y=0 donde se encuentra con el medio 
E = E0 
x = x0 
y = y0 
tita = tita0
```

```python
#COMIENZO DEL LOOP
for w in range (10000):  #particulas que tiro
    
    E = E0 
    x = x0 
    y = y0 
    tita = tita0
    
    for v in range (50): 

    ##distancia que viaja

        clm= E
        dist = - clm * np.log(np.random.random())

        x = dist * np.cos(tita) + x
        y = dist * np.sin(tita) + y

        if x<0.:
            k1 += 1
            break
        
        if abs(y)>120:
            break
        
        if x>=70:
            
            if np.abs(tita)>np.pi:
            
                extra=np.arcsin(np.sin(tita))
                tita_f= extra
            else:
                tita_f= tita
            
            angulo = np.int(tita_f*180/np.pi)
            vec_ang[angulo-90] +=1
           
            k2 += 1
            break
        
    ##Scattering
        sign = -1. 
        if np.random.random() < 0.5:
            sign = 1.
        
        tita = tita + sign *np.pi * (np.cos(np.pi*np.random.random()/2.))**2
```

```python
#si me interesa saber cuantas partículas salen del medio

print('Particulas que no atraviesan el slab (x<0):{}'.format(k1),"--->i.e:{}%".format(np.int(k1*100/(w+1))))
print('Particulas que atraviesan el slab:{}'.format(k2),"--->i.e:{}%".format(np.int(k2*100/(w+1))))


eje_ang = np.arange(181)-90

fig3= plt.subplots(figsize=(12,7))
plt.plot(eje_ang,vec_ang, '.', color='green' )
plt.title(u'Distribución angular', fontsize=14)

plt.show()
```
