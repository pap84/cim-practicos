## Ejercicio 1:

Implementar un método basado en Monte Carlo para determinar la probabilidad de obtener un “doble 6” lanzando 2 dados.


```python
n= 100

dado1 = np.random.randint(1,7,n) 
dado2 = np.random.randint(1,7,n) 

suma = dado1 + dado2

histograma = [[i,0] for i in range(1,max(suma) +1)]

for j in range(len(histograma)):
    for i in range(len(suma)):
        if suma[i] == histograma[j][0]:
            histograma[j][1] = histograma[j][1] + 1

x = []
y = []
for i in range(len(histograma)):
    x.append(histograma[i][0])
    y.append(histograma[i][1])
    
plt.bar(x,y)
plt.show()
```


## Ejercicio 2:

Calcular con método Monte Carlo el area debajo de la semilongitud de onda de una onda sinusoidal. Usando los **métodos de la media muestral y exito/fracaso**.  

*Voy a trabajar en un rectangulo con base $\pi$ y altura $1$. Cuya area sería igual a $\pi$.*
*Pasos a seguir: *
*
1. En el rectangulo defino la función con la que voy a trabajar
2. Tiro pares de numeros aleatorios para definir puntos aleatorios dentro del rectangulo (x,y)
3. Si los puntos estan debajo de la curva que define la función (f(x)) los cuento
4. Tomo una cantidad (N) suficiente de puntos (x,y)< (x,f(x)) son utilizados para calcular el Area usando el concepto de sumas de Riemman*

```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np
from time import time
```

```python
##METODO EXITO FRACASO
N = 10000
num = 0
g = np.zeros([N,2])

for w in range(N):
    x = np.pi*np.random.random_sample()
    y = np.random.random_sample()

    func_x = np.sin(x)
    
    if y<func_x:
        g[num,0] = x
        g[num,1] = y
        num = num + 1
        
#Calculo el area
A_calc = num/N *np.pi 

# se el area exacta es 2.
error = np.abs(2 - A_calc)

print ('Area obtenida: {}'.format(round(A_calc,4)), 'error: {}'.format(round(error, 4)))

#creo el seno para graficarlo:
base = np.linspace(0,3.14,num=1000)
plt.plot(base, np.sin(base), '-', color='red', label='sen(x)')
plt.plot(g[:,0], g[:,1],'g.', label='Método exito/fracaso')

plt.xlim(xmin=0, xmax=3.1415)
plt.title('Método del rechazo, N=1000, seno(x)', fontsize=17)
plt.legend()
plt.show()
```

```python
##METODO DE LA MEDIA MUESTRAL

N = 10000
suma = 0
num = 0

for w in range(N):
    x = np.pi*np.random.random_sample()
    funcion = np.sin(x)
    
    suma = suma + funcion
    
    g[num,0] = x
    g[num,1] = funcion
    
    num = num + 1
    
#Calculo el area
A_calc = suma /N *np.pi 

# se el area exacta es 2.
error = np.abs(2 - A_calc)

print ('Area obtenida: {}'.format(round(A_calc,4)), 'error: {}'.format(round(error, 4)))

#creo el seno para graficarlo:
plt.plot(g[:,0], g[:,1],'g.', label='metodo media muestral')
base = np.linspace(0,3.14,num=1000)
plt.plot(base, np.sin(base), '-', color='red', label='sen(x)')

plt.xlim(xmin=0, xmax=3.1415)
plt.title('Método del rechazo, N=1000, seno(x)', fontsize=17)
plt.legend()
plt.show()
```

## Ejercicio 4:
Calcular con método Monte Carlo la siguiente integral definida:
$$\int_{-1}^{1} (2 + u^{3/2}) du$$



*Calculo la integral entra -1 y 1, esto va a tener una parte imaginaria que no tengo en cuenta. Wolframalpha:*

$$\int_{-1}^{1} (2 + x^{3/2}) dx = \frac{22}{5} - \frac{2i}{5} \approx 4.4 - 0.4 i$$

*Como tomo la parte real deberia darme aprox 4.4*


```python
##METODO DE LA MEDIA MUESTRAL
N=10000
suma = 0
num = 0
f2 = np.zeros([N,2])

def funcion(w):
    f = 2 + w**(1.5)
    return f.real
    
for w in range(N):
    x = 2*np.random.random_sample()-1
    y = funcion(x) 
    
    suma = suma + y
    
    f2[num,0] = x
    f2[num,1] = y    
    num = num + 1
    
#Calculo el area
A_calc = suma /N * 2.

print ('Area obtenida media muest:{}'.format(round(A_calc,4)))

#creo el seno para graficarlo:


#creo el seno para graficarlo:
plt.plot(f2[:,0], f2[:,1],'b.', label='metodo media muestral')
base = np.linspace(0,3.14,num=10000)
plt.plot(base, funcion(base), '-', color='red', label='función a integrar')

plt.xlim(xmin=0, xmax=1)
plt.title('Integral de la funcion, N=10000', fontsize=17)
plt.legend()
plt.show()
```

$$ \int_{1/4}^{4} \frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}} dx$$

Estimar la desviación estándar en función del número de muestreo.
**PREGUNTAR**

*Recordar la desviación estandar puede calcularse como: *
$$ \sigma\ = \sqrt{\frac{1}{N}\left[ \left(\displaystyle\sum_{i=1}^{N}x_i^2\right)-NX^2 \right]} $$

*Siendo X el valor medio y $x_{i}$ el correspondiente al calculado en la iteración i*

```python
##METODO DE LA MEDIA MUESTRAL
N=10000
suma = 0
convergencia = np.zeros(int(N/20.)-1)
num = 0
i= 0
g2 = np.zeros([N,2])

def funcion(w):
    f = np.exp(-(w**2)/2.)
    return f
    
for w in range(N):
    x = 15/4.*np.random.random_sample()+ 1./4.
    y = funcion(x) 
    
    suma = suma + y

    if  (np.mod(w,20)==0 and w!=0):
        convergencia[i] = suma/(w *np.sqrt(2*np.pi)) *(15./4.) 
        i= i+1
    
    g2[num,0] = x
    g2[num,1] = y    
    num = num + 1

#Calculo el area [entre 1/16-16] y divido por raiz(2 pi)
A_calc = suma /(np.sqrt(2*np.pi)* N) *(15./4.)
error = np.abs(A_calc-0.34930)
eje_x = 20*np.arange(int(N/20.)-1)+20

print ('Valor integral definida:{}'.format(round(A_calc,4)),'Error:{}'.format(round(error,4)))
```

## Ejercicio 6:

Modelar el experimento de Buffon para estimar el valor de $\pi$, asignando valores que correspondan a los parámetros del problema.


*Los párametros que voy a asignar son los más simples posible: aguja y distancia entre lineas igual (D=1). Con estos valores puedo aproximar $\pi$ de la siguiente forma:*
$$ \pi \approx 2 \frac{N}{A}$$
*Siendo A la cantidad de aciertos si se hacen N intentos*


*Voy a tirar un número random entre 0 y 1 que define donde se encuentra el centro de la aguja (eje y)y otro número que va a determinar el ángulo (entre 0 y 90). Luego voy a contar si cruza o no la línea. Recordar la longitud de la guja es 1, de su centro tengo una distancia d=1/2*

```python
A = 0
N = 10000

for w in range(N):
    
    centro = np.random.random()
    angulo = np.random.random() * np.pi/2.
    y = np.sin(angulo)/2.

    if centro > 0.5 and (centro+y) >1:
        A += 1
    if centro < 0.5 and (centro-y) <0:
        A += 1
    
pi = 2*N/A
error = np.pi-pi

print('Valor aproximado de pi usando Buffon:{}'.format(round(pi,4)), 'error:{}'.format(round(error,4)))
```


## Ejercicio 7:

Realizar una simulación Monte Carlo simple para determinar la distancia total $D_T$ y distancia neta *D* (distancia al punto de partida) luego de 10, 100, 1000 y 10000 pasos de 1 unidad de una partícula moviéndose en un plano.

*Tengo N pasos por hacer, son todos de 1 unidad y no tengo la posibilidad de no hacer pasos, entonces: luego de N pasos la distancia total recorrida sí o sí es N.*
*Hago el gráfico para todos los casos.*

*Notar que defino el vector para N pasos de tamaño N+1 porque sino tengo un paso menos del que quiero. Así 10 pasos se necesita un vector con 11 elementos tomando el primero como el origen*

```python
#PASOS 10
x=0
y=0
N0= 10
vec_x = np.zeros(N0+1)
vec_y = np.zeros(N0+1)

for w in range(N0):
       
    angulo = 2.*np.pi*np.random.random() 
    x = x + np.cos(angulo)
    y = y + np.sin(angulo)  
    vec_x[w+1]=x
    vec_y[w+1]=y
    
d0 = np.sqrt(x**2 + y**2)

print('PASOS: {}'.format(N0))
print('Distancia neta: {}'. format(round(d0,2)), 'Posición final(x,y):({0},{1})'.format(round(x,2),round(y,2)))

plt.plot(vec_x , vec_y, '-o', label='camino recorrido')
plt.plot([0,x], [0,y] , '-o', color='red', label = 'distancia neta')
plt.grid(True)
plt.legend()
plt.show()
```

## Ejercicio 8: 

Repetir el ejercicio del item anterior para una partícula moviéndose en 3 dimensiones.

*Recordar en esfericas:*

$$ x = r \ sen(\theta) \ cos(\phi) \hspace{2cm} 
    y = r \ sen(\theta) \ sen(\phi) \hspace{2cm} z = r \ cos(\theta) $$
    
*En mi caso como quiero pasos de longitud unidad tomo r = 1*

*Para poder graficar en 3D debo agregar la linea: `ax = plt.axes(projection='3d')`*

```python
#PASOS 10
x=0
y=0
z=0
N0= 10
vec_x = np.zeros(N0+1)
vec_y = np.zeros(N0+1)
vec_z = np.zeros(N0+1)

for w in range(N0):
     
    tita = np.pi*np.random.random()        # tita entre [0,pi]    
    phi = 2.*np.pi*np.random.random()      # phi entre [0,2pi]
    x = x + np.sin(tita) * np.cos(phi)
    y = y + np.sin(tita) * np.sin(phi)  
    z = z + np.cos(tita)
    vec_x[w+1]=x
    vec_y[w+1]=y
    vec_z[w+1]=z
    
d0 = np.sqrt(x**2 + y**2 + z**2)

#GRAFICO
print('PASOS: {}'.format(N0))
print('Distancia neta: {}'. format(round(d0,2)), 'Posición final(x,y,z):({0},{1},{2})'.format(round(x,2),round(y,2),round(z,2)))

fig = plt.figure()
ax = plt.axes(projection='3d')

plt.plot(vec_x , vec_y, vec_z, '-o', label='camino recorrido')
plt.plot([0,x], [0,y] ,[0,z], '-o', color='red', label = 'distancia neta')
plt.grid(True)
plt.legend()
plt.title(u'Random walk, N=10', fontsize=12)
plt.show()
```

## Ejercicio 9:

Realizar una simulación Monte Carlo del transporte de partículas en 2D que sólo pueden interactuar de dos modos: absorción o scattering caracterizados por secciones eficaces $\sigma_A$ y $\sigma_S$ , respectivamente. En particular, se tiene que la distribución angular de $\sigma_S$ es isotrópica e independiente de la energía, mientras que $\sigma_A = C/E$ , donde *C* es una constante que satisface la normalización. El problema consiste en calcular la transmisividad de una muestra de espesor *L* por parte de una haz puntual de partículas de energía inicial $E_0$. Fijar valores de absorción (completo depósito local de la energía residual), de modo tal que no se extienda demasiado el tiempo de simulación.

*Sólo me interesa la transmividad de la muestra, no voy a hacer una matriz para ver el deposito de energía en la muestra, sino solo un gráfico que muestre la energía del haz que llega a salir de ella.*

*Los valores de C, $\sigma_S$ los elijo arbitrariamente. Tomo L = 70(espesor del Slab) *
*Los valores elegidos son de tal forma que aproximadamente un 60% de ellas atraviesan el material mientras que del resto 99% hacen backscaterring*

```python
#VALORES FIJOS
E0=50
C= 0.3 #valor arbitrario que elijo yo
sigma_s=0.01 #valor arbitrario que elijo yo (seccion eficaz scattering)
tol = 2.
L = 170.
N = 1000

vec_y = np.zeros(N) #se usa para graficar
vec_E = np.zeros(N) 

#Conteo
k1=0 #nro de particulas que salen por backscattering
k2=0 #nro de particulas que atraviesan el slab
k3=0 #nro de particulas absorbidas

for w in range (N):  #particulas que tiro
   
    #inicializacion variable
    E=E0
    x=0.
    y=0.
    tita = 0

    for v in range (1000): #seguimiento 1 particula
        
        sigma_a=C/E        #seccion eficaz absorcion
        sigma_t= sigma_s + sigma_a
        clm = 1./ sigma_t

    ##distancia que viaja
        dist = - clm * np.log(np.random.random())

        x = dist * np.cos(tita) + x
        y = dist * np.sin(tita) + y

        if x<0.:  #salio del medio
            k1 += 1
            break
    
        if x>L : #salio del slab
            delta_x = x - L
            y_salida = y - np.tan(tita)*delta_x
            vec_E[k2] = int(E)
            vec_y[k2] = int(y_salida)
            k2 += 1
            break
    
        if np.random.random()*sigma_t < sigma_a: #Absorcion
    
            aux = (np.random.random())**2
    
            if aux>0.95: #absorcion total
                E=0.
                k3 += 1
                break
            
            if E > E0/2.:
                E = E *(1. - aux/2.)
            
            else:
                E = E *(1. - aux) 
            
            if E<tol: #deposito lo que quedo en ese lugar
                k3 += 1    
                E=0.
                break        
    
        else: #scattering
    
            sign = -1. 
            if np.random.random() < 0.5:
                sign = 1.
         
            tita = tita + sign *np.pi*np.random.random()

print(u'Porcentaje de particulas que atraviesan el material:{}%'.format(k2*100/N))
print(u'Porcentaje de particulas que son absorbidas:{}%'.format(k3*100/N))
print(u'Porcentaje de particulas que realizan backscattering y salen del material:{}%'.format(k1*100/N))
```

```python
#Recorto los vectores:
vec_E = np.resize(vec_E, k2)
vec_y = np.resize(vec_y, k2)


#Espectro de energías:
Es = np.zeros(51)
for j in range (k2):
    for i in range(51):       
        if int(vec_E[j])==i:
            Es[i] += 1

#GRAFICO            
fig = plt.figure(figsize=(12,8))
plt.plot(Es, 'bo')
plt.grid(True)
plt.locator_params(axis='x', nbins=52)
plt.locator_params(axis='y', nbins=20)
plt.title(u'Espectro de energías', fontsize=14)
plt.show()
```

```python
#Posición final
y_salida = np.zeros(21)

for j in range (k2):
    for i in range(10):
        i=i+1
        if int(vec_y[j])==0:
            y_salida[10] += 1
        else:
            if abs(vec_y[j]) <= (i*50)   : 
                if vec_y[j]>0:
                    y_salida[10+i] += 1
                else:
                    y_salida[10-i] +=1
            
#GRAFICO     
axes_x = (np.arange(21)-10)*50
fig = plt.figure(figsize=(12,8))
plt.plot(axes_x, y_salida, 'bo')
plt.grid(True)
plt.locator_params(axis='x', nbins=22)
plt.locator_params(axis='y', nbins=10)
plt.title(u'Posicion de salida', fontsize=14)
plt.show()
```

En el último gráfico estan condensados los puntos:
